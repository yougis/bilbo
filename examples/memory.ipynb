```python
import ctypes
from dask.distributed import Client

```

```python
client = Client('172.20.12.13:9786')

def trim_memory() -> int:
    libc = ctypes.CDLL("libc.so.6")
    return libc.malloc_trim(0)

client.run(trim_memory, workers=['172.20.12.14:5004','172.20.12.17:5004','172.20.12.11:5004'])
```
