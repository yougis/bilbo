```python
client = Client('172.20.12.13:9786')

def trim_memory() -> int:
    libc = ctypes.CDLL("libc.so.6")
    return libc.malloc_trim(0)

client.run(trim_memory ,workers=['172.20.12.15:5004','172.20.12.14:5004','172.20.12.16:5004','172.20.12.17:5004','172.20.10.112:5004'])
```
