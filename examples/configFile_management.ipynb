```python
import os
import yaml

def ajouter_version_yaml(dossier):
    # Parcours de tous les fichiers dans le dossier
    for fichier in os.listdir(dossier):
        chemin_fichier = os.path.join(dossier, fichier)
        # Vérification si le fichier est un fichier YAML
        if os.path.isfile(chemin_fichier) and fichier.endswith('.yaml'):
            # Ouverture du fichier en mode lecture
            with open(chemin_fichier, 'r') as f:
                contenu = f.read()
            # Ajout de "version: 1" au début du fichier YAML
            contenu_modifie = "version: 1\n" + contenu
            # Réécriture du fichier avec la nouvelle version
            with open(chemin_fichier, 'w') as f:
                f.write(contenu_modifie)
            print(f"Le fichier {fichier} a été mis à jour avec succès.")
# Spécifiez le chemin du dossier contenant les fichiers YAML


# Appel de la fonction pour ajouter "version: 1" aux fichiers YAML dans le dossier spécifié
ajouter_version_yaml(dossier)
```

```python

def ajouter_libelle_yaml(dossier):
    # Parcours de tous les fichiers dans le dossier
    for fichier in os.listdir(dossier):
        chemin_fichier = os.path.join(dossier, fichier)
        # Vérification si le fichier est un fichier YAML
        if os.path.isfile(chemin_fichier) and fichier.endswith('.yaml'):
            # Ouverture du fichier YAML en mode lecture
            with open(chemin_fichier, 'r') as f:
                contenu = yaml.load(f, Loader=yaml.FullLoader)
            # Ajout de l'entrée "libelle" avec le nom du fichier comme valeur
            contenu['libelle'] = fichier[:-5]  # Exclure l'extension .yaml du nom du fichier
            # Réécriture du fichier avec la nouvelle entrée "libelle"
            with open(chemin_fichier, 'w') as f:
                yaml.dump(contenu, f)
            print(f"Le fichier {fichier} a été mis à jour avec succès.")

ajouter_libelle_yaml(dossier)
```
